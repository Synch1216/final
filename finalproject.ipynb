```python
import urllib2
import pandas as pd
from pandas import DataFrame
import json
```

```python
html = urllib2.urlopen('http://api.nytimes.com/svc/elections/us/v3/finances/2016/president/totals.json?api-key=fc8ac3e86a593008d6cbf415e8b23091:19:69821416')
```

```python
data=json.load(html)
data
```

```python


name= []
candidate_name= []
total_receipts= []
cash_on_hand= []

for value in data ['results']:
    name.append(value['name'])
    total_receipts.append(value['total_receipts'])
    #cycle.append(value['cycle'])
    cash_on_hand.append (value ['cash_on_hand'])

```

```python
election= zip(name, total_receipts, cash_on_hand)
df= DataFrame( data= election, columns= ['Name','Receipts','Cash On Hand'])
df
```

```python
df.plot(kind="bar", title="Funds")
```
